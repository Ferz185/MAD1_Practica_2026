**Capa oculta** (pesos $W^{(1)}$ y bias $b^{(1)}$):
$$W^{(1)} = \begin{pmatrix} 0.5 & -1.0 \\ 0.8 & 0.3 \end{pmatrix}, \quad b^{(1)} = \begin{pmatrix} 0.1 \\ -0.2 \end{pmatrix}$$

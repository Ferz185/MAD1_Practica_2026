## Parte 1 — Forward pass a mano

Considerá una red MLP con:
- **Capa de entrada:** 2 neuronas ($x_1$, $x_2$)
- **Capa oculta:** 2 neuronas con activación **ReLU** ($\max(0, z)$)
- **Capa de salida:** 1 neurona con activación **Sigmoide** ($\sigma(z) = \frac{1}{1+e^{-z}}$)

Los pesos y bias son:

**Capa oculta** (pesos $W^{(1)}$ y bias $b^{(1)}$):
$$W^{(1)} = \begin{pmatrix} 0.5 & -1.0 \\ 0.8 & 0.3 \end{pmatrix}, \quad b^{(1)} = \begin{pmatrix} 0.1 \\ -0.2 \end{pmatrix}$$

**Capa de salida** (pesos $W^{(2)}$ y bias $b^{(2)}$):
$$W^{(2)} = \begin{pmatrix} 1.2 & -0.7 \end{pmatrix}, \quad b^{(2)} = 0.0$$

La observación de entrada es $\mathbf{x} = (1.0,\; 2.0)$.

El forward pass calcula:
1. $\mathbf{z}^{(1)} = W^{(1)} \mathbf{x} + b^{(1)}$
2. $\mathbf{a}^{(1)} = \text{ReLU}(\mathbf{z}^{(1)})$
3. $z^{(2)} = W^{(2)} \mathbf{a}^{(1)} + b^{(2)}$
4. $\hat{y} = \sigma(z^{(2)})$

**Ejercicio 1.1** — Calculá $\mathbf{z}^{(1)}$ y $\mathbf{a}^{(1)}$ (capa oculta). Mostrá el cálculo paso a paso.

*Recordá: $\mathbf{z}^{(1)}_i = \sum_j W^{(1)}_{ij} x_j + b^{(1)}_i$*

$$\mathbf{z}^{(1)}=W^{(1)}\mathbf{x}+\mathbf{b}^{(1)}=
\begin{pmatrix}0.5 & -1.0\ 0.8 & 0.3\end{pmatrix}\begin{pmatrix}1\2\end{pmatrix}+\begin{pmatrix}0.1\-0.2\end{pmatrix}=\begin{pmatrix}-1.4\1.2\end{pmatrix}$$
Aplicando ReLU:
$$\mathbf{a}^{(1)}=\text{ReLU}(\mathbf{z}^{(1)})=\begin{pmatrix}0\1.2\end{pmatrix}$$

**Ejercicio 1.2** — Calculá $z^{(2)}$ y $\hat{y}$ (capa de salida).

Con $\mathbf{a}^{(1)}=(0,1.2)$:
$$z^{(2)}=W^{(2)}\mathbf{a}^{(1)}+b^{(2)}=(1.2,-0.7)\begin{pmatrix}0\1.2\end{pmatrix}+0=-0.84$$
Entonces:
$$\hat{y}=\sigma(z^{(2)})=\frac{1}{1+e^{0.84}}\approx 0.301$$

**Ejercicio 1.3** — Si la clase real es $y = 1$, calculá la pérdida BCE (Binary Cross-Entropy):

$$\mathcal{L} = -\left[ y \log(\hat{y}) + (1-y) \log(1-\hat{y}) \right]$$

Si $y=1$:
$$\mathcal{L}_{BCE}=-[y\log(\hat{y})+(1-y)\log(1-\hat{y})]=-\log(0.301)\approx 1.20$$

**Ejercicio 1.4** — Verificá tus cálculos con NumPy.

```python
import numpy as np

# Pesos y bias
W1 = np.array([[0.5, -1.0], [0.8, 0.3]])
b1 = np.array([0.1, -0.2])
W2 = np.array([[1.2, -0.7]])
b2 = np.array([0.0])

x = np.array([1.0, 2.0])
y_real = 1

# Forward pass — TU CÓDIGO ACÁ
z1 = W1@x+b1   # capa oculta: pre-activación
a1 = np.maximum(z1, 0)   # capa oculta: post-activación ReLU
z2 = W2@a1+b2   # capa de salida: pre-activación
y_hat = 1 / (1 + np.exp(-z2)) # capa de salida: sigmoide

# Pérdida BCE
loss = -(y_real * np.log(y_hat) + (1 - y_real) * np.log(1 - y_hat))  # TU CÓDIGO ACÁ

print(f'z1     = {z1}')
print(f'a1     = {a1}')
print(f'z2     = {z2}')
print(f'y_hat  = {y_hat[0]:.4f}')
print(f'BCE    = {loss[0]:.4f}')
```

## Parte 2 — MLP con scikit-learn

```python
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

X, y = make_moons(n_samples=500, noise=0.25, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)
```

**Ejercicio 2.1** — Entrenás una MLP con una capa oculta de 32 neuronas, activación ReLU y optimizador Adam. Completá el código.

```python
mlp = MLPClassifier(
    hidden_layer_sizes=(32,),   # TU CÓDIGO ACÁ: (32,)
    activation='relu',            # TU CÓDIGO ACÁ: 'relu'
    solver='adam',                # TU CÓDIGO ACÁ: 'adam'
    max_iter=500,
    random_state=42
)
mlp.fit(X_train_sc, y_train)  # TU CÓDIGO ACÁ

print(f"Accuracy test: {accuracy_score(y_test, mlp.predict(X_test_sc)):.3f}")
```

**Ejercicio 2.2** — Graficá la curva de pérdida durante el entrenamiento.

```python
fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(mlp.loss_curve_)
ax.set_xlabel('Iteración')
ax.set_ylabel('Pérdida (BCE)')
ax.set_title('Curva de pérdida — MLP')
plt.tight_layout()
plt.show()
```

**Ejercicio 2.3** — Comparás arquitecturas: (16,), (32,), (64,), (32, 16), (64, 32). Para cada una, calculás el accuracy en test.

```python
arquitecturas = [(16,), (32,), (64,), (32, 16), (64, 32)]

for arch in arquitecturas:
    mlp_a = MLPClassifier(
        hidden_layer_sizes=arch,  # TU CÓDIGO ACÁ
        activation='relu', solver='adam',
        max_iter=500, random_state=42
    )
    mlp_a.fit(X_train_sc, y_train)           # TU CÓDIGO ACÁ
    acc = accuracy_score(y_test, mlp_a.predict(X_test_sc))  # TU CÓDIGO ACÁ
    print(f'Arquitectura {str(arch):12s} — Accuracy test: {acc:.3f}')
```

**Ejercicio 2.4** — ¿Más neuronas/capas siempre mejora el resultado? ¿Qué riesgo aparece al aumentar la capacidad de la red?

No, sumar neuronas/capas no mejora automáticamente. En este experimento, pasar de (16,) a (64,) mejor?, pero hacer la red más profunda no dio mejoras consistentes.

El riesgo al aumentar demasiado capacidad es sobreajuste.

**Ejercicio 2.5** — Entrenás la misma MLP con SGD (Stochastic Gradient Descent) y comparás con Adam.

```python
mlp_sgd = MLPClassifier(
    hidden_layer_sizes=(32,), activation='relu',
    solver='sgd',   # TU CÓDIGO ACÁ: 'sgd'
    learning_rate_init=0.01,
    max_iter=500, random_state=42
)
mlp_sgd.fit(X_train_sc, y_train)

fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(mlp.loss_curve_,     label='Adam')
ax.plot(mlp_sgd.loss_curve_, label='SGD')
ax.set_xlabel('Iteración')
ax.set_ylabel('Pérdida')
ax.set_title('Adam vs SGD — curvas de pérdida')
ax.legend()
plt.tight_layout()
plt.show()
```

**Ejercicio 2.6** — ¿Qué diferencia observás en la curva de pérdida entre Adam y SGD? ¿Cuál converge más rápido?

En las curvas de este ejercicio, SGD cae más rápido al inicio y Adam desciende de forma más gradual. Esa relación puede cambiar según learning rate y demás hiperparámetros.

No hay optimizador universalmente mejor: depende de configuración y problema.

## Preguntas de reflexión final

1. En el forward pass a mano, una neurona oculta con z negativo fue anulada por ReLU. ¿Qué implicancia tiene eso para el aprendizaje?
2. ¿Por qué es importante escalar entradas antes de entrenar una MLP?
3. Una red con muchas capas/neuronas puede memorizar el entrenamiento. ¿Cómo podés limitar eso sin reducir drásticamente la capacidad?

1. Si ReLU anula una neurona (salida 0), su gradiente local puede quedar en cero en ese paso, frenando su actualización temporalmente.

2. Escalar entradas ayuda a estabilizar la optimización, mejora convergencia y evita que variables con mayor escala dominen el gradiente.

3. Para limitar memorizar: regularización (L2), dropout, early stopping, más datos y validación cuidadosa de hiperparámetros.
